```python
def stress(w=1):
    sigma = lambda t: np.sin(w*t)/(w*t)
    dsigma = lambda t: (1/t)*(np.cos(w*t) - np.sin(w*t)/(w*t))
    
    return (sigma, dsigma)
```

```python
def strain(w=1):
    epsilon = lambda t: np.sin(w*t)/(w*t)
    depsilon = lambda t: (1/t)*(np.cos(w*t) - np.sin(w*t)/(w*t))
    
    return (epsilon, depsilon)
```

```python
def creep(elastic_moduli, viscosities):
    E = np.array(elastic_moduli).reshape(-1, 1)
    tau = np.array(viscosities).reshape(-1, 1) / E[1:, :]
    
    J = lambda t: 1/ E[0, :] + np.sum((1 - np.exp(- t / tau)) / E[1:, :])
    
    return J
```

```python
def relax(elastic_moduli, viscosities):
    E = np.array(elastic_moduli).reshape(-1, 1)
    tau = np.array(viscosities).reshape(-1, 1) / E[1:, :]
    
    R = lambda t: E[0, :] + np.sum(np.exp(-t / tau) * E[1:, :])
    
    return R
```

```python
def calculate_strain_single_t(stress_funcs, creep_func, dt,time_array):
    sigma, dsigma = stress_funcs
    integrand = lambda t_p: creep_func(dt - t_p) * dsigma(t_p)
    strain_single_t = sigma(time_array[0]) * creep_func(dt) + quad(integrand, time_array[0], dt)[0]
    return strain_single_t
```

```python
def calculate_stress_single_t(strain_funcs, relax_func, dt, start_time):
    epsilon, depsilon = strain_funcs
    integrand = lambda t_p: relax_func(dt - t_p) * depsilon(t_p)
    stress_single_t = epsilon(start_time) * relax_func(dt-start_time) + quad(integrand, start_time, dt)[0]
    return stress_single_t
```

```python
def calculate_strain(time_array, E_mods, viscs, freq=1):    
    stress_funcs = stress(w=freq)
    creep_func = creep(E_mods, viscs)
    strain = np.array([calculate_strain_single_t(stress_funcs, creep_func, dt, time_array) for dt in time_array])
    return strain
```

```python
def calculate_stress(time_array, E_mods, viscs, freq=1):    
    strain_funcs = strain(w=freq)
    relax_func = relax(E_mods, viscs)
    stress = np.array([calculate_stress_single_t(strain_funcs, relax_func, dt, time_array[0]) for dt in time_array])
    return stress
```

### Calculate library

```python
def num_derivs(dependent_data, independent_data, diff_order):
    
    data_derivs = dependent_data.copy()
    data_derivs = data_derivs.reshape(-1, 1)
    for _ in range(diff_order):
        data_derivs = np.append(data_derivs, np.gradient(data_derivs[:, -1].flatten(), independent_data.flatten()).reshape(-1,1), axis=1)
    
    return data_derivs
```

```python
diff_order = len(viscs)

strain_theta = num_derivs(strain_array, time_array, diff_order)
stress_theta = num_derivs(stress_array, time_array, diff_order)
num_theta = np.append(strain_theta, stress_theta, axis=1)
num_theta.shape
```

Conclusion: Least squares method was actually less cumbersome and much faster. Also, solve doesn't work, even after lots of thinking! The thing is, lstsqs jumps straight to a numerical approach to this kind of problem. sympy.solve eventually will attempt some similar things, but it tries first many other, more analytical methods. I think when you have a system of 5000 simultaneous equations, sym.solve is too clever and slow an algorithm. Also, it will only be satisfied if there is a pefect solution. with 5000 equations and not totally accurate (rounding errors, numerical derivatives) values, no perfect solution will exist.
